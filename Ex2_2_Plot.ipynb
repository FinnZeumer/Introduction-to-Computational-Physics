```python
import numpy as np
import Papulator as pap
from Papulator import const
from Papulator import Colors as c
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

z_val = np.array([40, 41, 42, 44, 45, 46, 47])

def cal_frq(k, Z, sigma):
    return k*(Z-sigma)

# c/ lam und daraus die Wurzel ziehen 10^- 9cm 
lam_val = np.array([6.091, 5.749, 5.423, 4.861, 4.622, 4.385, 4.170])

freq_val = np.sqrt(const.c / lam_val)


plt.scatter(z_val, freq_val, label='Datapoints')

def linear(a,x,b):
    return a * x + b

x = np.linspace(39, 48, 1000)

popt, pcov = curve_fit(linear, z_val, freq_val)


plt.plot(x, linear(popt[0], x, popt[1]), label='Lineare Regressionsfunktion', color = c.BLUE)
plt.vlines(43, 7.25e3, 8e3, color = c.WINE_RED, label='Missing Value')
plt.text(39, 8e3, fr'$\lambda_{{43}}$ = {const.c / (linear(popt[0], 43, popt[1]))**2:.3f} ')
pap.plot_me('Finden eines längst vergessenen Elementes (es war nicht bekannt)', 'Z', r'$\sqrt{\nu}$', safe_as='Ex4_H2_A2_PLT')

print(linear(popt[0], 43, popt[1]))
```
